```python
import csv
```

```python
query_names = {}
with open("./renameQ.csv", 'r') as file:
    csvreader = csv.reader(file)
    i = 0
    for row in csvreader:
        if i == 0:
            i += 1
            continue
        query_names[row[0].split('/')[1]] = row[1][:-4]
```

```python
ref_names = {}
with open("./rename.csv", 'r') as file:
    csvreader = csv.reader(file)
    i = 0
    for row in csvreader:
        if i == 0:
            i += 1
            continue
        ref_names['/'.join(row[0].split('/')[1:])] = row[1][:-4]
```

```python
gt = []
```

```python
oxford_landmarks = ['all_souls', 'ashmolean', 'balliol', 'bodleian', 'christ_church', 'cornmarket', 'hertford', 'keble',
                    'magdalen', 'pitt_rivers', 'radcliffe_camera']
paris_landmarks = ['defense', 'eiffel', 'invalides', 'louvre', 'moulinrouge', 'museedorsay', 'notredame', 'pantheon', 'pompidou', 'sacrecoeur', 'triomphe']
```

```python
queries = []
for landmark in oxford_landmarks:
    for i in range(1, 6):
        query_file = f"{landmark}_{str(i)}_query.txt"
        with open(f"/home/augustinas/OxParis/oxford_gt/{query_file}") as my_file:
            query = f"{my_file.readline().split(' ')[0][5:]}.jpg"
            queries.append(query)
            print(query)
for landmark in paris_landmarks:
    for i in range(1, 6):
        query_file = f"{landmark}_{str(i)}_query.txt"
        with open(f"/home/augustinas/OxParis/paris_gt/{query_file}") as my_file:
            query = f"{my_file.readline().split(' ')[0]}.jpg"
            queries.append(query)
            print(query)
```

```python
gt=[]
for i in range(len(oxford_landmarks)):
    landmark = oxford_landmarks[i]
    for j in range(1, 6):
        good_file = f"{landmark}_{str(j)}_good.txt"
        with open(f"/home/augustinas/OxParis/oxford_gt/{good_file}") as my_file:
            Lines = my_file.readlines()
            for line in Lines:
                if f"{line[:-1]}.jpg" in queries:
                    continue
                gt.append([query_names[queries[i*5+j-1]], ref_names[f"{line[:-1]}.jpg"]])
                print([query_names[queries[i*5+j-1]], ref_names[f"{line[:-1]}.jpg"]])
    for j in range(1, 6):
        ok_file = f"{landmark}_{str(j)}_ok.txt"
        with open(f"/home/augustinas/OxParis/oxford_gt/{ok_file}") as my_file:
            Lines = my_file.readlines()
            for line in Lines:
                if f"{line[:-1]}.jpg" in queries:
                    continue
                gt.append([query_names[queries[i*5+j-1]], ref_names[f"{line[:-1]}.jpg"]])
                print([query_names[queries[i*5+j-1]], ref_names[f"{line[:-1]}.jpg"]])
```

```python
for i in range(len(paris_landmarks)):
    landmark = paris_landmarks[i]
    for j in range(1, 6):
        good_file = f"{landmark}_{str(j)}_good.txt"
        with open(f"/home/augustinas/OxParis/paris_gt/{good_file}") as my_file:
            Lines = my_file.readlines()
            for line in Lines:
                if f"{line[:-1]}.jpg" in queries:
                    continue
                gt.append([query_names[queries[55+i*5+j-1]], ref_names[f"{line[:-1]}.jpg"]])
                print([query_names[queries[55+i*5+j-1]], ref_names[f"{line[:-1]}.jpg"]])
    for j in range(1, 6):
        ok_file = f"{landmark}_{str(j)}_ok.txt"
        with open(f"/home/augustinas/OxParis/paris_gt/{ok_file}") as my_file:
            Lines = my_file.readlines()
            for line in Lines:
                if f"{line[:-1]}.jpg" in queries:
                    continue
                gt.append([query_names[queries[55+i*5+j-1]], ref_names[f"{line.split('_')[1]}/{line[:-1]}.jpg"]])
                print([query_names[queries[55+i*5+j-1]], ref_names[f"{line.split('_')[1]}/{line[:-1]}.jpg"]])
```

```python
gt.sort(key=lambda x: x[0])
```

```python
with open('gt.csv', 'w') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerow(['query_id', 'reference_id'])
    for row in gt:
        csvwriter.writerow(row)
```
