```python
import argparse
from typing import Optional

import matplotlib.pyplot as plt
import numpy as np

from isc.io import read_ground_truth, read_predictions
from isc.metrics import evaluate, Metrics, print_metrics

import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import numpy as np
```

```python
d_precision = []
d_recall = []
d_racall_10 = []
for i in range(1, 26):
    predictions = read_predictions(f"./data/glv2_t/disc/{i}/matrix_no_norm.csv")
    gts = read_ground_truth('./data/glv2_t/ground_truth.csv')
    metrics = evaluate(gts, predictions)
    d_precision.append(metrics.average_precision)
    d_recall.append(metrics.recall_at_rank1)
    d_racall_10.append(metrics.recall_at_rank10)
    if metrics.recall_at_p90 is None:
        print(f"| {str(i).zfill(2)}            | {metrics.average_precision:.5f}               | -                 | -                   | {metrics.recall_at_rank1:.5f}              | {metrics.recall_at_rank10:.5f}               |")
    else:
        print(f"| {str(i).zfill(2)}            | {metrics.average_precision:.5f}               | {metrics.recall_at_p90:.5f}           | {metrics.threshold_at_p90:g}             | {metrics.recall_at_rank1:.5f}              | {metrics.recall_at_rank10:.5f}               |")
```

```python
q_precision = []
q_recall = []
q_racall_10 = []
for i in range(1, 26):
    predictions = read_predictions(f"./data/glv2_q/glv2_q/{i}/matrix_no_norm.csv")
    gts = read_ground_truth('./data/glv2_q/ground_truth.csv')
    metrics = evaluate(gts, predictions)
    q_precision.append(metrics.average_precision)
    q_recall.append(metrics.recall_at_rank1)
    q_racall_10.append(metrics.recall_at_rank10)
    if metrics.recall_at_p90 is None:
        print(f"| {str(i).zfill(2)}            | {metrics.average_precision:.5f}               | -                 | -                   | {metrics.recall_at_rank1:.5f}              | {metrics.recall_at_rank10:.5f}               |")
    else:
        print(f"| {str(i).zfill(2)}            | {metrics.average_precision:.5f}               | {metrics.recall_at_p90:.5f}           | {metrics.threshold_at_p90:g}             | {metrics.recall_at_rank1:.5f}              | {metrics.recall_at_rank10:.5f}               |")
```

```python
t_precision = []
t_recall = []
t_racall_10 = []
for i in range(1, 26):
    predictions = read_predictions(f"./data/glv2_t/glv2_t/{i}/matrix_no_norm.csv")
    gts = read_ground_truth('./data/glv2_t/ground_truth.csv')
    metrics = evaluate(gts, predictions)
    t_precision.append(metrics.average_precision)
    t_recall.append(metrics.recall_at_rank1)
    t_racall_10.append(metrics.recall_at_rank10)
    if metrics.recall_at_p90 is None:
        print(f"| {str(i).zfill(2)}            | {metrics.average_precision:.5f}               | -                 | -                   | {metrics.recall_at_rank1:.5f}              | {metrics.recall_at_rank10:.5f}               |")
    else:
        print(f"| {str(i).zfill(2)}            | {metrics.average_precision:.5f}               | {metrics.recall_at_p90:.5f}           | {metrics.threshold_at_p90:g}             | {metrics.recall_at_rank1:.5f}              | {metrics.recall_at_rank10:.5f}               |")
```

```python
epochs = np.arange(1, 26, 1)
plt.plot(epochs, d_precision, 'o-r', label='ViT-D')
plt.plot(epochs, t_precision, 'o-g', label='ViT-LT')
plt.plot(epochs, q_precision, 'o-b', label='ViT-LQ')
plt.xlim(1, 25)
plt.ylim(bottom=0, top=None)
plt.xticks([1, 5, 10, 15, 20, 25])
plt.xlabel("Epocha")
plt.legend()
# plt.savefig('./glv2_q_loss.eps', format='eps')
plt.show()
```

```python
epochs = np.arange(1, 26, 1)
plt.plot(epochs, q_precision, 'o-r', label='P')
plt.plot(epochs, q_recall, 'o-g', label='R@1')
plt.plot(epochs, q_racall_10, 'o-b', label='R@10')
plt.xlim(1, 25)
plt.ylim(bottom=0, top=None)
plt.xticks([1, 5, 10, 15, 20, 25])
plt.xlabel("Epocha")
plt.legend(loc='upper right')
# plt.savefig('./glv2_q_loss.eps', format='eps')
plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, sharex=True, sharey=True, layout='compressed', figsize=(6, 4))

plt.xlim(1, 25)
plt.xticks([1, 5, 10, 15, 20, 25])
ax2.set_xlabel("Epocha")

ax1.plot(epochs, d_precision, 'r', label='μAP')
ax1.plot(epochs, d_recall, 'g', label='R@1')
ax1.plot(epochs, d_racall_10, 'b', label='R@10')
ax1.set_title('ViT-D')

ax2.plot(epochs, t_precision, 'r')
ax2.plot(epochs, t_recall, 'g')
ax2.plot(epochs, t_racall_10, 'b')
ax2.set_title('ViT-LT')

ax3.plot(epochs, q_precision, 'r')
ax3.plot(epochs, q_recall, 'g')
ax3.plot(epochs, q_racall_10, 'b')
ax3.set_title('ViT-LQ')

fig.legend(loc='outside upper right')
plt.savefig('./land_vit_metrics.eps', format='eps')
plt.show()
```
