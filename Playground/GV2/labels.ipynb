```python
import csv
from tqdm.auto import tqdm
import random
```

```python
images_to_classes = {}
index = 0
with open("./train_clean_5000.csv", 'r') as file:
    csvreader = csv.reader(file)
    for row in tqdm(csvreader, desc="Reading"):
        index += 1
        if index == 1:
            continue
        images_to_classes[row[0]] = row[1]
```

```python
images_to_classes
```

```python
renames = {}
index = 0
with open("./rename.csv", 'r') as file:
    csvreader = csv.reader(file)
    for row in tqdm(csvreader, desc="Reading"):
        index += 1
        if index == 1:
            continue
        renames[row[0]] = row[1]
```

```python
renames
```

```python
with open("./labels.csv", 'w') as file:
    csvwriter = csv.writer(file)
    csvwriter.writerow(["image", "landmark_id"])
    for image, landmark in images_to_classes.items():
        csvwriter.writerow([renames[image], landmark])
```
